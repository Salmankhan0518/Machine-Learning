## Working with Date

```python
# Converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
date.info()
```

## Extract Year

```python
date['date_year'] = date['date'].dt.year
date.head()
```

## Extract Month

```python
date['date_month_no'] = date['date'].dt.month
date.head()
```

```python
# If you want name of the month
date['date_month_no'] = date['date'].dt.month_name()
date.head()
```

## Extract Days

```python
date['date_day'] = date['date'].dt.day
date.head()
```

```python
# if you want day name of the month
date['date_day'] = date['date'].dt.day_name()
date.head()
```

### Day of the week

```python
date['day_now'] = date['date'].dt.dayofweek
date.head()
```

```python
date['day_now_name'] = date['date'].dt.day_name()
date.head()
```

## Checking is weekend?

```python
date['date_is_weekend'] = np.where(date['day_now_name'].isin(['Saturday', 'Sunday']), 1,0)
date.drop(columns=['product_id', 'city_id', 'orders']).head()
```

## Extract week of the year

```python
date['date_week'] = date['date'].dt.isocalendar().week
date.head()
```

### Extract Quater

```python
date['quater'] = date['date'].dt.quarter
date.head()
```

## Extract Semester

```python
date['semester'] = np.where(date['quater'].isin([1,2]), 1, 2)
date.drop(columns=['product_id', 'city_id', 'orders']).head()
```

## Extract Time elapsed between dates

```python
import datetime

today = datetime.datetime.today()

today
```

```python
today - date['date']
```

```python
# if only want days
(today - date['date']).dt.days
```

```python
# month passed
np.round((today - date['date']).dt.days /30.41, 0)
```
