## EDA Univariate Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Titanic dataset

```python
df = pd.read_csv("train.csv")
df
```

## 1. Categorical Data
### a. Count Chart

```python
#sns.countplot(x='Survived', data=df, color='g')
#sns.countplot(x='Pclass', data=df, color='g')
sns.countplot(x='Sex', data=df, color='g')
#print(df['Survived'].value_counts().plot(kind='bar'))
plt.grid()
plt.show()
```

## b. PieChart

```python
print(df['Sex'].value_counts().plot(kind='pie', autopct='%.1f'))
plt.show()
```

## 2. Numerical Data

## a. Histogram

```python
plt.hist(df['Age'], bins=50)
plt.grid()
plt.show()
```

## b. Box Plot

```python
sns.boxplot(x='Age', data=df)
```

```python
sns.distplot(df['Age'])
plt.show()
```
