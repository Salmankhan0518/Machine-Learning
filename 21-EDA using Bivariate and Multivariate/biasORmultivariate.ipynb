```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
titanic = pd.read_csv('train.csv')
titanic
```

```python
flights = pd.read_csv('flights.csv')
flights[['year', 'month', 'hour']]
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

## 1. Scatter Plot(Numerical-Numerical)

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex', style=tips['smoker'], size=tips['size'])
plt.show()
```

## 2. Bar Plot(Numerical-Categorical)

```python
titanic.head()
```

```python
sns.barplot(x='Pclass', y='Age', data=titanic, hue='Sex')
```

## 3. Box Plot(Numerical - Categorical)

```python
sns.boxplot(x='Sex', y='Age', data=titanic, hue='Survived')
```

## 4. Distplot(Numerical - Categorical)

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'])
sns.distplot(titanic[titanic['Survived']==1]['Age'])
plt.show()
```

## 5. HeatMap(Categorical - Categorical)

```python
pd.crosstab(titanic['Pclass'], titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
plt.show()
```

```python
titanic.groupby('Sex')['Survived'].mean() * 100
```

```python
titanic.groupby('Pclass')['Survived'].mean() * 100
```

```python
(titanic.groupby('Pclass')['Survived'].mean() * 100).plot(kind='bar')
plt.show()
```

## 6. Cluster Map(Categorical - Categorical)

```python
pd.crosstab(titanic['SibSp'], titanic['Survived'])
```

```python
sns.clustermap(pd.crosstab(titanic['Parch'], titanic['Survived']))
plt.show()
```
