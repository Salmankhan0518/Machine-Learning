### 3. Opening a csv file from an URL

```python
import requests 
from io import StringIO

url = "https://gist.githubusercontent.com/rnirmal/e01acfdaf54a6f9b24e91ba4cae63518/raw/6b589a5c5a851711e20c5eb28f9d54742d1fe2dc/datasets.csv"

headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.8) Gecko/20100101 Firefox/66.0"}

req = requests.get(url, headers=headers)

data = StringIO(req.text)

pd.read_csv(data)
```

### 14. Convertor

```python
def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    else:
        name
```

```python
rename("Royal Challengers Bangalore")
```

```python
pd.read_csv("IPL Matches 2008-2020.csv", converters= {'team1': rename})
```
