```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df['number'].unique()
```

```python
fig = df['number'].value_counts().plot.bar()
fig.set_title('Passengers travelling with')
```

## Extract Numerical part

```python
df['number_numerical'] = pd.to_numeric(df['number'], errors='coerce', downcast='integer')
df['number_catrgorical'] = np.where(df['number_numerical'].isnull(), df['number'],np.nan)

df['number_catrgorical'] = df['number_catrgorical'].fillna('Missing')
df['number_numerical'] = df['number_numerical'].fillna(0)

df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python
df['cabin_num'] = df['Cabin'].str.extract('(\d+)')
df['cabin_cat'] = df['Cabin'].str[0]

df['cabin_num'] = df['cabin_num'].fillna(0)
df['cabin_cat'] = df['cabin_cat'].fillna('Missing')

df.head()
```

```python
df['cabin_cat'].value_counts().plot(kind='bar')
```

### Extract the last bit of the ticket number

```python
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],
                                errors='coerce',
                                downcast='integer')

# extract the first part of the ticket category
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan, df['ticket_cat'])

df.head()
```
